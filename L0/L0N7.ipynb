```python
from mpl_toolkits.mplot3d import Axes3D

from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

# определение данных
A = 10
pi = 3.1415
X = np.arange(-5.12, 5.12, 0.25)
Y = np.arange(-5.12, 5.12, 0.25)
X, Y = np.meshgrid(X, Y)
Z = 2*A+(X*X-A*np.cos(2/pi*X))+(Y*Y-A*np.cos(2/pi*Y))

# график
surf = ax.plot_surface(X, Y, Z)

# линии уровня
fig = plt.figure()
ax = fig.gca()

# построение линий уровня
cs = ax.contour(X, Y, Z)

# добавление подписей к линиям уровня
ax.clabel(cs)

plt.show()
```
