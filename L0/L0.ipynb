```python
from mpl_toolkits.mplot3d import Axes3D

from matplotlib import cm
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()
ax = fig.gca(projection='3d')

#  
X = np.arange(-100, 100, 0.1)
Y = np.arange(-100, 100, 0.1)
X, Y = np.meshgrid(X, Y)
Z = -20*2.71828**(-0.2**((1/2)*(0.5*((X**2)+(Y**2)))))-2.71828**(0.5*np.cos(2*np.pi*X)+np.cos(2*np.pi*Y))+20+2.71828

# 
surf = ax.plot_surface(X, Y, Z)

#  
fig = plt.figure()
ax = fig.gca()

#   
cs = ax.contour(X, Y, Z)

#     
ax.clabel(cs)

plt.show()
```
